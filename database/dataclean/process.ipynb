```python
## read in products from amazon_products.csv into a dataframe
import pandas as pd
df = pd.read_csv("amazon_products.csv")

#read in categories from amazon_categories.csv into a dataframe
df_cat = pd.read_csv("amazon_categories.csv")
```

```python
print(df.columns)
```

```python
#keep columns title, imgURL, price, category_id
df = df[['title', 'imgUrl', 'price', 'category_id']]
df.head()

df_cat.head()
```

```python
#now corroborate category_id with category name. category_id is called id in df_cat     
df = pd.merge(df, df_cat, how='left', left_on='category_id', right_on='id')
df.head()
```

```python
# First, get the top 20 categories by frequency
top_20_categories = df['category_name'].value_counts().nlargest(20).index.tolist()

# Then, sample 100 products from each of these top 20 categories
df_top20 = df[df['category_name'].isin(top_20_categories)].groupby('category_name').apply(lambda x: x.sample(n=100, replace=True)).reset_index(drop=True)
```

```python
import random #change title to name, change price to base_price, change category_name to category, add a description column, add a stock column with a random number from 0-1000, 
df_top20['name'] = df_top20['title']
df_top20['base_price'] = df_top20['price']
df_top20['category'] = df_top20['category_name']
df_top20['description'] = df_top20['title']
df_top20['stock'] = df_top20['title'].apply(lambda x: random.randint(0, 1000))
df_top20.head()
```

```python
#delete the columns title, price, category_id, id
df_top20 = df_top20.drop(columns=['title', 'price', 'category_id', 'id'])
df_top20.head()
```
